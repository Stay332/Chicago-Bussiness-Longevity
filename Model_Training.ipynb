### Q 1.2


 Prepare the data for analysis by performing the necessary cleaning, transformation, and preprocessing steps.

```python
chicago_licenses_df_sub = chicago_licenses_df[['CITY', 'ZIP CODE', 'LICENSE STATUS', 'LICENSE TERM START DATE','LICENSE TERM EXPIRATION DATE','LATITUDE', 'LONGITUDE', 'ACCOUNT NUMBER', 'LICENSE DESCRIPTION', "LEGAL NAME",
                                               "ADDRESS"]]
chicago_licenses_df_sub = chicago_licenses_df_sub[chicago_licenses_df_sub["LICENSE STATUS"].isin(["AAI", "AAC"])].reset_index(drop=True)
```

```python
chicago_licenses_df_sub['LICENSE DESCRIPTION'].value_counts(dropna=False).head(10)
```

```python
import pandas as pd

current_observation_date = pd.Timestamp.now()

chicago_licenses_df_sub['START_DATE'] = pd.to_datetime(chicago_licenses_df_sub['LICENSE TERM START DATE'], errors='coerce')
chicago_licenses_df_sub['EXP_DATE']   = pd.to_datetime(chicago_licenses_df_sub['LICENSE TERM EXPIRATION DATE'], errors='coerce')
chicago_licenses_df_sub['EXP_DATE']   = chicago_licenses_df_sub['EXP_DATE'].fillna(current_observation_date)
chicago_licenses_df_sub = chicago_licenses_df_sub.dropna(subset=['START_DATE'])

chicago_licenses_df_sub = chicago_licenses_df_sub.sort_values(by='EXP_DATE')

top_10_licenses = chicago_licenses_df_sub['LICENSE DESCRIPTION'].value_counts().nlargest(10).index.tolist()
chicago_licenses_df_sub = chicago_licenses_df_sub[chicago_licenses_df_sub['LICENSE DESCRIPTION'].isin(top_10_licenses)]

df_merged = chicago_licenses_df_sub.groupby('ACCOUNT NUMBER').agg(
    MIN_START_DATE=('START_DATE',        'min'),
    MAX_EXP_DATE  =('EXP_DATE',          'max'),
    LATEST_STATUS =('LICENSE STATUS',    'last'),
    LATITUDE      =('LATITUDE',          'first'),
    LONGITUDE     =('LONGITUDE',         'first'),
    ZIP_CODE      =('ZIP CODE',          'first'),
    LEGAL_NAME    =('LEGAL NAME',        'first'),
    ADDRESS       =('ADDRESS',           'first'),
    ALL_LICENSES  =('LICENSE DESCRIPTION', lambda x: list(set(x.dropna())))
).reset_index()

def determine_real_event(row):
    if row['LATEST_STATUS'] == 'AAC':
        return 1
    elif row['MAX_EXP_DATE'] < current_observation_date:
        return 1
    else:
        return 0

df_merged['event']    = df_merged.apply(determine_real_event, axis=1)
df_merged['duration'] = (df_merged['MAX_EXP_DATE'] - df_merged['MIN_START_DATE']).dt.days / 365.25
df_merged = df_merged[df_merged['duration'] > 0]
```

```python
df_merged.dropna(inplace=True)
df_merged["ZIP_CODE"] = df_merged["ZIP_CODE"].astype(int)
```

### Q 1.3 — Enrich with ZIP Code Features (Rent & Consumer Spending)

Add two socioeconomic features from the **2022 ACS 5-year estimates** (U.S. Census Bureau):

| Feature | ACS Variable | Interpretation |
|---|---|---|
| `MEDIAN_RENT` | B25064_001E — Median Gross Rent ($/month) | Proxy for local commercial rent pressure |
| `MEDIAN_INCOME` | B19013_001E — Median Household Income ($/yr) | Proxy for local consumer spending power |

Both are merged at the **ZIP code** level via a left join on `ZIP_CODE`.


```python
import requests

# ── Fetch 2022 ACS 5-year estimates for all US ZCTAs, then filter ──────────
# B25064_001E : Median Gross Rent ($/month)     → proxy for local rent level
# B19013_001E : Median Household Income ($/yr)  → proxy for resident spending power
#
# NOTE: The Census API does NOT support &in=state:XX for ZCTAs because ZIP codes
# cross state boundaries. We fetch all ~33k ZCTAs nationally and filter below.
ACS_URL = (
    "https://api.census.gov/data/2022/acs/acs5"
    "?get=B25064_001E,B19013_001E"
    "&for=zip%20code%20tabulation%20area:*"
)

print("Fetching ACS data from Census Bureau (all US ZCTAs)…")
resp = requests.get(ACS_URL, timeout=60)
resp.raise_for_status()

header, *rows = resp.json()
acs_df = pd.DataFrame(rows, columns=header).rename(columns={
    "B25064_001E":              "MEDIAN_RENT",
    "B19013_001E":              "MEDIAN_INCOME",
    "zip code tabulation area": "ZIP_CODE",
})

acs_df["ZIP_CODE"]      = acs_df["ZIP_CODE"].astype(int)
acs_df["MEDIAN_RENT"]   = pd.to_numeric(acs_df["MEDIAN_RENT"],   errors="coerce")
acs_df["MEDIAN_INCOME"] = pd.to_numeric(acs_df["MEDIAN_INCOME"], errors="coerce")

# Census uses large negative sentinels (-666666666, -999999999) for "not available"
acs_df.loc[acs_df["MEDIAN_RENT"]   < 0, "MEDIAN_RENT"]   = None
acs_df.loc[acs_df["MEDIAN_INCOME"] < 0, "MEDIAN_INCOME"] = None

# Keep only ZIPs that appear in our dataset
chicago_zips = set(df_merged["ZIP_CODE"].unique())
acs_df = acs_df[acs_df["ZIP_CODE"].isin(chicago_zips)][["ZIP_CODE", "MEDIAN_RENT", "MEDIAN_INCOME"]]

print(f"Matched {len(acs_df):,} ZIP codes from our dataset")
acs_df.describe().round(0)
```

```python
# ── Merge ACS features into df_merged on ZIP_CODE ─────────────────────────
df_merged = df_merged.merge(acs_df, on="ZIP_CODE", how="left")

# Fill the few unmatched ZIPs with the Illinois statewide median
df_merged["MEDIAN_RENT"].fillna(df_merged["MEDIAN_RENT"].median(), inplace=True)
df_merged["MEDIAN_INCOME"].fillna(df_merged["MEDIAN_INCOME"].median(), inplace=True)

print(f"Shape after enrichment: {df_merged.shape}")
print("\nNew feature statistics:")
df_merged[["ZIP_CODE", "MEDIAN_RENT", "MEDIAN_INCOME"]].describe().round(0)
```

```python
import pandas as pd
from lifelines import CoxPHFitter
from sklearn.model_selection import train_test_split

# One-Hot Encoding for ALL_LICENSES
licenses_dummies = df_merged['ALL_LICENSES'].str.join('|').str.get_dummies()

# Build feature matrix — now includes MEDIAN_RENT and MEDIAN_INCOME
df_model = pd.concat([
    df_merged[['LATITUDE', 'LONGITUDE', 'MEDIAN_RENT', 'MEDIAN_INCOME', 'duration', 'event']],
    licenses_dummies
], axis=1)

# 80/20 split
train_df, test_df = train_test_split(df_model, test_size=0.2, random_state=42)

# Cox PH model with L2 penalizer to handle collinearity
cph = CoxPHFitter(penalizer=0.1)

print("Start training Cox Proportional Hazards model...")
cph.fit(train_df, duration_col='duration', event_col='event')

train_score = cph.score(train_df, scoring_method='concordance_index')
test_score  = cph.score(test_df,  scoring_method='concordance_index')

print(f"\nTraining completed!")
print(f"Training C-index: {train_score:.4f}")
print(f"Testing  C-index: {test_score:.4f}")

print("\n--- Model Feature Weights Summary ---")
cph.print_summary()
```

```python
import pickle
import pandas as pd

with open('chicago_business_survival_model.pkl', 'wb') as f:
    pickle.dump(cph, f)
    
print("✅ Successfully saved chicago_business_survival_model.pkl")
with open('chicago_business_survival_model.pkl', 'rb') as f:
    loaded_cph = pickle.load(f)

print("✅ Successfully loaded the model! Ready for predictions.")
```

```python
required_features = list(loaded_cph.params_.index)
new_business_data = {feature: 0 for feature in required_features}

# Location: The Loop (ZIP 60601)
new_business_data['LATITUDE']  = 41.881832
new_business_data['LONGITUDE'] = -87.623177

# ACS values for ZIP 60601 (approximate 2022 estimates)
new_business_data['MEDIAN_RENT']   = 2450   # $/month
new_business_data['MEDIAN_INCOME'] = 112000 # $/year

# License type
new_business_data['Retail Food Establishment'] = 1

new_df = pd.DataFrame([new_business_data])
survival_curve = loaded_cph.predict_survival_function(new_df)

def get_prob_at_year(curve, target_year):
    nearest_idx = curve.index.get_indexer([target_year], method='nearest')[0]
    return curve.iloc[nearest_idx].values[0]

print("\n--- Business Longevity Forecast ---")
print(f"Location : The Loop, ZIP 60601")
print(f"Median Rent   : $2,450/month")
print(f"Median Income : $112,000/year")
print(f"License Type  : Retail Food Establishment")
print()
for yr in [1, 3, 5, 10]:
    print(f"Year {yr:2d}: {get_prob_at_year(survival_curve, yr):.2%}")
```
